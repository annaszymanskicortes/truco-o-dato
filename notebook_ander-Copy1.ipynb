```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1617 = pd.read_csv('data_Ander/data_promotions_1617.csv', encoding='latin1')
data1617 = data1617[data1617['Municipi'] == 'BARCELONA']
data1617 = data1617.drop(columns = ['Nivell', 'Sexe'])
col_resultat = 'Resultat de l\'avaluació'
data1617['Promociona'] = np.where(
    data1617[col_resultat].str.startswith('Promo'),
    'PROMOCIONA',
    'NO PROMOCIONA'
)
col_n_alumnes = "Nombre d'avaluats"
resumen = (data1617
    .groupby('Promociona')[col_n_alumnes].sum().reset_index()
)

resumen
resumen['Proporció'] = resumen["Nombre d'avaluats"] / resumen["Nombre d'avaluats"].sum()

resumen1617 = resumen
resumen1617
```

```python
data1718 = pd.read_csv('data_Ander/data_promotions_1718.csv')
data1718 = data1718[data1718['Municipi'] == 'BARCELONA']

col_resultat = "Resultat de l'avaluació"

data1718 = data1718[data1718[col_resultat] != "Alumnes avaluats"]

data1718['Promociona'] = np.where(
    data1718[col_resultat].str.startswith('Promo'),
    'PROMOCIONA',
    'NO PROMOCIONA'
)
col_n_alumnes = "Nombre d'avaluats"
resumen1718 = (data1718
    .groupby('Promociona')[col_n_alumnes].sum().reset_index()
)

resumen1718['Proporció'] = resumen1718["Nombre d'avaluats"] / resumen1718["Nombre d'avaluats"].sum()

resumen1718 
```
